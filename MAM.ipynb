#### oś OX

 $$\mathbf{R}_x(\alpha)  = \begin{bmatrix} 1 & 0 & 0 \\ 0 & \cos \alpha & -\sin \alpha \\ 0 & \sin \alpha & \cos \alpha\end{bmatrix}$$

###### Wyznacznik

$$\det \mathbf{R}_x(\alpha) = 1\cdot \cos^2\alpha - \left(1\cdot(-\sin^2\alpha)\right) = \cos^2\alpha + \sin^2\alpha = 1$$

##### Wartości własne

$$w_{\mathbf{R}_x(\alpha)}(\lambda) = \det\left(\mathbf{R}_x(\alpha) - \lambda I\right) = (1-\lambda)
(\cos\alpha - \lambda)^2 + (1-\lambda)\sin^2\alpha = (1-\lambda)(\cos^2\alpha -2\lambda\cos\alpha + \lambda^2) + (1-\lambda)\sin^2 \alpha = (1-\lambda)(\lambda^2-2\cos\alpha \cdot \lambda +1)$$

$$\Delta = 4\cos^2\alpha - 4 \leqslant 0$$, czyli tylko dla $\cos^2\alpha =1$ będą rzeczywiste rozwiązania (to jest obrót o kąt półpełny, czyli _de facto_ inwersja. Wówczas wartość własna jest zdegenerowana i rónwna $-1$. A wówczas wektory własne będą wektorami z bazy, czyli $\begin{bmatrix} 1 \\ 0 \\ 0\end{bmatrix}$, $\begin{bmatrix} 0 \\ 1 \\ 0\end{bmatrix}$, $\begin{bmatrix} 0 \\ 0 \\ 1\end{bmatrix}$, bo współrzędna _x_-owa się nie zmieni (wartość własna 1), _y_- i _z_-ową na przeciwne (wartość własna -1).

w ogólności rozwiązania są takie $\frac{2\cos\alpha\pm2\mathrm{i}\sqrt{1-\cos^2\alpha}}{2} = \cos\alpha \pm \mathrm{i}\sin\alpha = e^{\pm \mathrm{i} \alpha}$

### pozostałe osie

$$\mathbf{R}_y(\alpha) = \begin{bmatrix} \cos\alpha & 0 & -\sin\alpha\\0 & 1 & 0 \\ \sin\alpha & 0 &\cos\alpha \end{bmatrix}$$

$$\mathbf{R}_z(\alpha) = \begin{bmatrix} \cos\alpha & -\sin\alpha & 0 \\ \sin\alpha & \cos\alpha & 0 \\ 0 & 0 & 1\end{bmatrix}$$

Rachunki analogiczne do pozostałych ($\det \mathbf{R}_i = 1$) i tak samo wartości własne. Ale wartość własna 1 ma wektor własny skierowane zgodnie z osią obrotu, a w przypadku rzeczywistych pierwiastków $-1$ pozostałe wektory bazy.

```julia
function A(a, b)
    [a -b;
     b a]
end
```

```julia
function B(n, A)
    B = [0 0; 0 0]
    for i in 1:n
        B = B*B + A
    end
    return B
end
```

```julia
kolory = []

for α in -1.99:0.01:1.99
    for β in -1.99:0.01:1.99
        if det(B(100, A(α, β))) < 2
                append!(kolory, [100*α, 100*β, "#000000"])
            else
                append!(kolory, [100*α, 100*β, "#FFFFFF"])
            end
        end
    end
```
